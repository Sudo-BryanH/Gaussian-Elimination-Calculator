### Code for the algorithm

```python
# collen is number of columns in matrix
# rowlen is number of rows in matrix
#testers 
# collen = 3
# rowlen = 2
# matrix = [[0.0, 1.0, 4.0], [2.0, 2.0, 3.0], [3.2, 0.0, 2.5], [2.0, 4.5, 6.2]]

collen = 3
rowlen = 4
matrix = [[1.0, 4.0, 6.0], [7.0, 8.0, 5.0], [69.0, 46.0, 7.0], [12.0, 4.0, 7.0]]

# matrix = new_mat

def printMat():
    for rw in matrix:
        print(rw)
    print(" ")


# takes the integers of the indexes of the rows and swaps the content
# requires r to be in the array
def rowSwap(r1, r2):
    # r1 = r1-1
    # r2 = r2-1
    temp = matrix[r1]
    matrix[r1] = matrix[r2]
    matrix[r2] = temp

# multiplies the row at index r by a scalar c
# requires r to be in the array
def scalarMultiply(c, r):
    # r = r-1
    rw = []
    for n in matrix[r] :
        b = n*c
        rw.append(b)
        
    matrix[r] = rw
        
        
# adds a multiple of row r1 to row r2
def rowReplace(r1, r2, c):
    # r1 = r1-1
    # r2 = r2-1
    rw = []
    for i in range(collen):
        row1 = (matrix[r1])
        row2 = (matrix[r2])
        n1 = row1[i]
        n2 = row2[i]
        rw.append(c*n1 + n2)
        
    
    matrix[r2] = rw

    
    
# matrix[row#][col#]
    
# Use a column tracker to avoid diiding by 0 in the case the whole column is 0
def rowReduce():
    currCol = 0
    printMat()

    for rw in range(0, rowlen):
        currCol = 0
        allZero = checkSwap(currCol, rw)
        while not allZero:
            currCol = currCol + 1
            allZero = checkSwap(currCol, rw)
        divisor = matrix[rw][currCol]
        neutralized = [row/divisor for row in matrix[rw]]
        print(f"scalar multiply row{rw+1} by {divisor}")
        matrix[rw] = neutralized
        printMat()
        for i in range(rw + 1, rowlen):
            print(matrix[i][currCol])
            print(matrix[rw][currCol])
            c = -matrix[i][currCol]/matrix[rw][currCol]      ## TODO: double check how a matrix should be represented
            rowReplace(rw, i, c)
            print(f"row replace row{i + 1} + ({c})*row{rw + 1}")
            printMat()
            
        currCol = currCol + 1
        if currCol > collen :
            break


        
# baserow
# returns if a column is all zero. Swaps br (current row) with first non-zero first row and returns true, else false
def checkSwap(col, br):
    swapped = False
    for i in range(br, rowlen):
        if matrix[i][col] != 0:
            rowSwap(br, i)
            if (i != br):
                print(f"swapped row{br + 1} and row{i + 1}")
                printMat()
            swapped = True
            break
            

            
    return swapped     

        
rowReduce()
```
